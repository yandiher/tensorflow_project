```python
data = data.drop('id', axis=1)
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data['month'] = data['date'].apply(lambda date: date.month)
data['year'] = data['date'].apply(lambda date: date.year)
data.drop('date', axis=1, inplace=True)
```

```python
data.groupby('month').mean()['price'].plot()
```

```python
data.head()
```

```python
data.drop('zipcode', axis=1, inplace=True)
```

```python
data.head()
```

```python
data['yr_renovated'].value_counts()
```

```python
data['sqft_basement'].value_counts()
```

```python
x = data.drop('price', axis=1).values
y = data['price'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=x_train,y=y_train, validation_data=(x_test,y_test),
          batch_size=128, epochs=400)
```

```python
loss = pd.DataFrame(model.history.history)
loss.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, r2_score
```

```python
prediction = model.predict(x_test)
```

```python
print(mean_squared_error(y_true=y_test,y_pred=prediction,squared=False))
print(mean_absolute_error(y_true=y_test,y_pred=prediction))
print(explained_variance_score(y_true=y_test,y_pred=prediction))
print(r2_score(y_true=y_test,y_pred=prediction))
```
